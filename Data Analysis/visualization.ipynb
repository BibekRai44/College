```python
#importing necessary module

import pandas as pd
import numpy as np
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#reading dataset

data=pd.read_csv("/Users/bibekrai/workspace/College/Data.csv")
```

```python
data=data.drop(columns={'Unnamed: 0'})
```

```python
data['University'].value_counts().shape
```

```python
#piechart

mylabels=data['University'].value_counts().head(5).index.to_list()
plt.pie(data=data,x=data['University'].value_counts().head(5),labels=mylabels,autopct='%1.1f%%')
plt.title("Percentage shared by Top 5 University",color='darkgreen')
```

```python
mylabels=data['Ownership Type'].value_counts().index.to_list()
plt.pie(data=data,x=data['Ownership Type'].value_counts().head(5),labels=mylabels,autopct='%1.1f%%')
plt.title("Percentage of Ownership type")
```

```python
mylabels=data['Course Offered'].value_counts().head(5).index.to_list()
myexplode=[0,0,0.5,0,0.5]
plt.pie(data=data,x=data['Course Offered'].value_counts().head(5),labels=mylabels,autopct='%1.1f%%',explode=myexplode)
plt.title("Top Course Provided by University")
```

```python
sns.scatterplot(data=data,x=data['Ownership Type'].head(10),y=data['Location'].head(10))
plt.title('Ownership type College based on Location',color='darkgreen')
```
